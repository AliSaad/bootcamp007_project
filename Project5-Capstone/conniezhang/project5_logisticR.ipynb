```python
from __future__ import print_function
import sklearn.metrics
import time

def evaluate_model(clf):
    """Scores a model using log loss with the created train and test sets."""
    start = time.time()
    clf.fit(x_tm9, y_tm9)
    train_loss = sklearn.metrics.log_loss(y_tm9, clf.predict_proba(x_tm9))
    test_loss = sklearn.metrics.log_loss(y_test9, clf.predict_proba(x_test9))
    print("Train score:", train_loss)
    print("Test score:", test_loss)
    print("Total time:", time.time() - start)
    print()
    return test_loss
```

```python
# Grid search: try 8 different possibilities
from sklearn.grid_search import GridSearchCV

params = {'C' : [1,100,10000], 'penalty': ['l1', 'l2'], 
          'class_weight':[{0:0.11}, {1:0.89}] }
grid = GridSearchCV(logReg9, params, scoring='log_loss')
evaluate_model(grid)
grid.grid_scores_
```
