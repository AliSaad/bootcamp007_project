```python
def sum_of_squares(user_1, user_2, reviews):
    import math
    import pandas as pd
    user_1Restaurants = reviews['business_id'].loc[reviews['user_id'] == user_1].tolist()
    user_2Restaurants = reviews['business_id'].loc[reviews['user_id'] == user_2].tolist()
    commonRestaurants = list(set(user_2Restaurants).intersection(set(user_1Restaurants)))
    if len(commonRestaurants) == 0:
        return []
    sumer = 1/(1+sum([math.sqrt(pow((reviews[stars].loc[(reviews[userId] == user_1) &\
                                                   (reviews['business_id'] == item)]).iloc[0] - \
                               (reviews[stars].loc[(reviews[userId] == user_2) & \
                                                   (reviews['business_id'] == item)]).iloc[0],2)) \
                 for item in commonRestaurants]))
    return [sumer , user_2, len(commonRestaurants)]
```
