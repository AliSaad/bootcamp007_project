```python
import sqlite3
conn = sqlite3.connect('/home/Chuan/PycharmProjects/foodAnalysis/tweets.db')
c = conn.cursor()
```

```python
c.execute('SELECT geo, text FROM tweetTable')
tweet_raw = c.fetchall()
tweets = filter(lambda x: x[0] != u'None', tweet_raw)
```

```python
import json
latlng = map(lambda x: json.loads(x[0].replace("'", "\"").replace('u', ''))[u'coordinates'], tweets)
texts = map(lambda x: x[1], tweets)
```

```python
print len(latlng)
print len(texts)
```

```python
import folium
from folium.plugins import MarkerCluster
import random

latlng_sample, text_sample = zip(*random.sample(zip(latlng, texts), 1000))

tweet_map = folium.Map(location=[30, 0], zoom_start=2)
tweet_map.add_children(MarkerCluster(locations=latlng_sample, 
                                     popups=text_sample))
```
