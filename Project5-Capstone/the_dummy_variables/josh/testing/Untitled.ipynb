```python
# mAP vs Rank
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

rank = [10, 20, 30, 40, 50]

mAP= [0.042818143,
0.072108896,
0.071031999,
0.065745802,
0.089934819]

pop = 0.017911323770145694
fontsize=14
c = 'white'

fig = plt.figure()
ax = fig.add_subplot(111)
ax.tick_params(axis='x', colors=c)
ax.tick_params(axis='y', colors=c)

plt.plot(rank, mAP, label="Implicit")
plt.ylabel('mAP', fontsize=fontsize, color=c)
plt.xlabel('Rank', fontsize=fontsize, color=c)
plt.ylim(0, 0.1)
plt.title('mAP vs. Rank', fontsize=fontsize, color=c)
plt.plot((10, 50), (pop, pop), 'k-', label="Popularity")
plt.legend(fontsize=fontsize, loc=2)


plt.show()
```

```python
# mAP vs Lambda
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

mAP = [0.096713279,
0.090388704,
0.100701518,
0.095902699,
0.083854326]

lambda_ = [0,
0.01,
0.1,
1,
10]

pop = 0.017911323770145694
fontsize=14
c = 'white'

fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_xscale('log')
ax.tick_params(axis='x', colors=c)
ax.tick_params(axis='y', colors=c)

plt.plot(lambda_, mAP, label="Implicit")
plt.ylabel('mAP', fontsize=fontsize, color=c)
plt.xlabel('Lambda', fontsize=fontsize, color=c)
plt.title('mAP vs. Lambda', fontsize=fontsize, color=c)
plt.plot((0, 10), (pop, pop), 'k-', label="Popularity")
plt.legend(fontsize=fontsize, loc=1)


plt.show()
```
