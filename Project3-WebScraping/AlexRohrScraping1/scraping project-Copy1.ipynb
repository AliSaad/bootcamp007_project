```python
#urls
base_url = "https://newyork.craigslist.org/search/rea"
urls = [base_url]
for i in range(1,25):
    urls.append(base_url + '?s=' + str(i * 100)) 

prices = []
hoods =[]
sqft = []
title = []

#loop over urls
for url in urls:
    r = urllib.urlopen(url).read()
    soup = BeautifulSoup(r)
    result_raw = soup.find_all('span', class_="result-meta")
    result_name = soup.find_all('a', class_="result-title hdrlnk")
    
    #loop over listings on page
    for i in range(0,100):
    #first we look at the prices and then after that we look at the hood 
    #prices
        if len (result_raw[i].find_all('span', class_="result-price")) > 0:
            prices.append(result_raw[i].find_all('span', class_="result-price")[0].getText())
        else:
            prices.append("NA")   
    #hood
        if len (result_raw[i].find_all('span', class_="result-hood")) > 0:
            hoods.append(result_raw[i].find_all('span', class_="result-hood")[0].getText())
        else:
            hoods.append("NA")   
    #sqft
        if len (result_raw[i].find_all('span', class_="housing")) > 0:
            sqft.append(result_raw[i].find_all('span', class_="housing")[0].getText())
        else:
            sqft.append("NA") 
    #title
        if len(result_name[i]) > 0:
            title.append(result_name[i].getText())
        else:
            title.append("NA") 
```
